```python
#Importing the needful libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

%matplotlib inline
```

```python
#creating my own data using pandas DataFrame(2-dimensional) datatype

Students = pd.DataFrame({"Name":["Benjamin","Angelina","Eileen","Rexford","Sandrina","Charlotte","Derrick","Emmanuella","Hayford","Robert"],
                        "DOB":["3/20/1996","5/10/1992","11/11/1991","1/08/1998","5/15/1956","3/20/2000","3/15/2005","8/20/1996","3/13/1993","12/12/2000"],
                        "Student_Id":["ITC0075","ITC0085","ITC0057","ITC0001","ITC0046","ITC0051","ITC0004","ITC0098","ITC0065","ITC0048"]})
#outputing the data

Students

#outputing the datatype of data

Students.dtypes
```

```python
#changing the datatype of the DOB column
Students["DOB"] = pd.to_datetime(Students["DOB"])

Students
Students.dtypes
```

```python
#Extracting the Month,Day,Year,'Day_of_week', 'Week_of_year' from the data

Students["Month"] = Students.DOB.dt.month
Students["Day"] = Students.DOB.dt.day
Students["Year"] = Students.DOB.dt.year
Students["Week Number"] = Students.DOB.dt.isocalendar().week
Students["Day Of Week"] = Students.DOB.dt.dayofweek
Students["Day Name"] = pd.to_datetime(Students["DOB"]).dt.day_name()
Students["Month Name"] = pd.to_datetime(Students["DOB"]).dt.month_name()

Students
```

```python
Students["Gender"] = ["M","F","F","M","F","F","M","F","M","M"]
```
